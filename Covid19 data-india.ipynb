```python
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup 
import requests
```

```python
website_url = requests.get('https://covid19ind.zaoapp.net/').text
```

```python
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
Title = soup.title.text
Title
```

```python
data_tag = soup.find_all('td')
data_tag
```

```python
State_Name = []
Confirmed  = []
Recovered = []
Deaths = []

    
```

```python
State_Name = []
for item in data_tag[0::4]:
    if len(item)==1:
        match = item.text

        State_Name.append(match)
    else:
        State_Name.append(np.nan)
```

```python
len(State_Name)
```

```python
Total_confirmed = []
for item in data_tag[1::4]:
    if len(item)==1:
        match = item.text

        Total_confirmed.append(match)
    else:
        Total_confirmed.append(np.nan)
```

```python
len(Total_confirmed)
```

```python
Recovered = []
for item in data_tag[2::4]:
    if len(item)==1:
        match = item.text

        Recovered.append(match)
    else:
        Recovered.append(np.nan)
```

```python
len(Recovered)
```

```python
Deaths = []
for item in data_tag[3::4]:
    if len(item)==1:
        match = item.text

        Deaths.append(match)
    else:
        Deaths.append(np.nan)
```

```python
len(Deaths)
```

```python
Covid_ind = pd.DataFrame()
Covid_ind['State_Name'] = State_Name
Covid_ind['Total_confirmed'] = Total_confirmed
Covid_ind['Total_Recovered'] = Recovered
Covid_ind['Total_Deaths'] = Deaths
```

```python
Covid_ind
```

```python
Covid_ind = Covid_ind.iloc[0:27,:]
```
